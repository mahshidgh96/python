## Visualization

```python
import matplotlib
import matplotlib.pyplot as plt

# Set the backend explicitly for Jupyter
matplotlib.use('module://matplotlib_inline.backend_inline')

medal_counts = [df['Gold'].sum(),df['Silver'].sum(),df['Bronze'].sum()]
medal_labels = ['Gold','Silver','Bronze']
plt.pie(medal_counts, labels=medal_labels, autopct='%.1f%%')
plt.title('Distribution by Olympics 2024')
plt.show()
```
